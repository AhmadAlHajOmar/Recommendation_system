```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = products_df.select_dtypes(include=['float64', 'int64']).columns

cleaned_df = products_df.copy()

outlier_mask = pd.Series([False] * len(cleaned_df))

for col in numeric_cols:
    Q1 = cleaned_df[col].quantile(0.25)
    Q3 = cleaned_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    is_outlier = (cleaned_df[col] < lower_bound) | (cleaned_df[col] > upper_bound)
    
    outlier_mask = outlier_mask | is_outlier

def remove_all_outliers(df, cols):
    df_cleaned = df.copy().reset_index(drop=True)
    while True:
        outlier_mask = pd.Series([False] * len(df_cleaned))
        for col in cols:
            Q1 = df_cleaned[col].quantile(0.25)
            Q3 = df_cleaned[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR
            outlier_mask |= (df_cleaned[col] < lower) | (df_cleaned[col] > upper)
        if not outlier_mask.any():
            break
        df_cleaned = df_cleaned[~outlier_mask].reset_index(drop=True)
    return df_cleaned

products_df = remove_all_outliers(products_df, numeric_cols)

print(f" Number of rows after removing all outliers: {len(products_df)}")
```

```python
numeric_cols = products_df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    data = products_df[col]
    
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    outliers = data[(data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))]
    
    if not outliers.empty:
        print(f" Column '{col}' has {len(outliers)} outliers.")
    else:
        print(f" Column '{col}' has no outliers.")
    
    plt.figure(figsize=(5, 2.5))
    sns.boxplot(x=data, color="skyblue")
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

<h1>Association Rules</h1>

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

product_id_to_name = products_df.set_index('ProductID')['ProductName'].to_dict()

transactions = invoices_df.groupby('InvoiceID')['ProductID'].apply(
    lambda ids: [product_id_to_name[pid] for pid in ids if pid in product_id_to_name]
).tolist()

te = TransactionEncoder()
te_array = te.fit_transform(transactions)
basket = pd.DataFrame(te_array, columns=te.columns_)

frequent_itemsets = apriori(basket, min_support=0.04, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

sorted_rules = rules.sort_values(by='lift', ascending=False)

for idx, rule in sorted_rules.iterrows():
    antecedents = ', '.join(rule['antecedents'])
    consequents = ', '.join(rule['consequents'])
    support = rule['support']
    confidence = rule['confidence']
    lift = rule['lift']

    print(f" Rule: ({antecedents}) => ({consequents})")
    print(f"   Support: {support:.2f}")
    print(f"   Confidence: {confidence:.2f}")
    print(f"   Lift: {lift:.2f}")
    print("-" * 40)
```

```python
sorted_rules = rules.sort_values(by='lift', ascending=False)
print(sorted_rules)
```

```python
confident_rules = rules[rules['confidence'] >= 0.75]
print(confident_rules)
```

<h1>Clustering</h1>

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
text_columns =['ProductName', 'Category','Brand', 'SupplierCountry','ConnectivityType', 'MaterialType','UsageType', 'PriceCategory']
label_encoders = {}
for col in text_columns: 
    le = LabelEncoder()
    products_df[col] = le.fit_transform(products_df[col])
    label_encoders[col] = le
display(products_df.head())
```

```python
features_with_price = products_df[['Price', 'Rating', 'Stock', 'WarrantyYears', 
                          'WeightKg', 'VolumeCm3', 'PowerWatt'] + text_columns]
scaler = StandardScaler()
scaled_with_price = scaler.fit_transform(features_with_price)
scaled_df_with_price = pd.DataFrame(scaled_with_price, columns=features_with_price.columns)
display(scaled_df_with_price.head())
```

```python
features_without_price  = products_df[[ 'Rating', 'Stock', 'WarrantyYears', 
                          'WeightKg', 'VolumeCm3', 'PowerWatt'] + text_columns]
scaler = StandardScaler()
scaled_without_price = scaler.fit_transform(features_without_price)
scaled_df_without_price = pd.DataFrame(scaled_without_price, columns=features_without_price.columns)
display(scaled_df_without_price.head())
```

```python

inertia_with_price = []


K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_with_price)
    inertia_with_price.append(kmeans.inertia_)
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(K_range, inertia_with_price, 'bo-')
plt.title('Elbow Method ( )')
plt.xlabel('  (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
inertia_without_price = []
K_range = range(1, 11)


for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_without_price)
    inertia_without_price.append(kmeans.inertia_)
plt.figure(figsize=(12,5))
plt.subplot(1,2,2)
plt.plot(K_range, inertia_without_price, 'go-')
plt.title('Elbow Method ( )')
plt.xlabel('  (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
print(" K_range:", len(K_range))
print(" inertia_without_price:", len(inertia_without_price))
print(" inertia_with_price:", len(inertia_with_price))
```

```python


def get_recommended_products(product_id, cluster_type='with_price'):
    """
         
    :param product_id:   
    :param cluster_type: 'with_price'  'without_price'
    :return: DataFrame  
    """
    #      
    cluster_col = 'Cluster_With_Price' if cluster_type == 'with_price' else 'Cluster_Without_Price'
    
    #    
    product_cluster = products_df.loc[products_df['ProductID'] == product_id, cluster_col].values[0]
    
    #          
    recommended = products_df[(products_df[cluster_col] == product_cluster) & (products_df['ProductID'] != product_id)]
    
    return recommended[['ProductID', 'ProductName', 'Price', 'Brand', 'Category']]

#    :
recommended_products = get_recommended_products(5, cluster_type='without_price')
display(recommended_products)
```

```python
import numpy as np

def evaluate_similarity(products_df, cluster_type='with_price', sample_size=3):
    """
             .
    :param products_df: DataFrame 
    :param cluster_type: 'with_price'  'without_price'
    :param sample_size:     
    :return:   
    """
    cluster_col = 'Cluster_With_Price' if cluster_type == 'with_price' else 'Cluster_Without_Price'
    
    scores = []
    
    #   
    sample_products = products_df.sample(n=sample_size, random_state=42)
    
    for _, product in sample_products.iterrows():
        product_id = product['ProductID']
        print(f"\n  : {product['ProductName']} (ID: {product_id}) | : {product['Price']}")
        
        #      
        recommended = products_df[(products_df[cluster_col] == product[cluster_col]) & (products_df['ProductID'] != product_id)]
        
        if recommended.empty:
            print("         .")
            continue
        
        for _, rec in recommended.iterrows():
            print(f"  : {rec['ProductName']} | : {rec['Price']}")
            while True:
                score_input = input("   (0-100)  'x' : ")
                if score_input.strip().lower() == 'x':
                    print("    .")
                    mean_score = np.mean(scores) if scores else 0
                    print(f"\n   ({cluster_type}): {mean_score:.2f}")
                    return scores, mean_score
                try:
                    score = int(score_input)
                    if 0 <= score <= 100:
                        scores.append(score)
                        break
                    else:
                        print("    0  100 .")
                except:
                    print("   .")
    
    #  
    mean_score = np.mean(scores) if scores else 0
    print(f"\n   ({cluster_type}): {mean_score:.2f}")
    
    return scores, mean_score


#      
scores_with_price, mean_with_price = evaluate_similarity(products_df, cluster_type='with_price', sample_size=2)

#      
scores_without_price, mean_without_price = evaluate_similarity(products_df, cluster_type='without_price', sample_size=2)

#    
print("\n    ")
if mean_with_price > mean_without_price:
    print(" **:**     .")
elif mean_with_price < mean_without_price:
    print(" **:**      .")
else:
    print(" **:**        .")
```
